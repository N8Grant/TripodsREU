## Split data into differnt classes

```python
def return_subgroups(gene, p, r):
    subgroups = {}
    patients = p.columns[2:]
    val = p.loc[[gene]].values[0,2:]
    for i in range(len(val)):
        if val[i] in subgroups:
            subgroups[val[i]].append((i,patients[i],val[i]))
        else:
            subgroups[val[i]] = [(i,patients[i],val[i])]
    return subgroups
```

```python
subgroups = return_subgroups("BAP1", presence_df,results_df)
```

```python
print(subgroups.keys())
```

```python
group_1_df = results_df.loc[[t[1] for t in subgroups[-1]]]
group_2_df = results_df.loc[[t[1] for t in subgroups[0]]]
group_1_df = group_1_df[group_1_df.columns.difference(['RMSE','P-value','Pearson Correlation'])]
group_2_df = group_2_df[group_2_df.columns.difference(['RMSE','P-value','Pearson Correlation'])]
```

## Plot KDE of groups distributions and run non-parametric t-test

```python
expression_df = pd.read_excel('C:/Users/NathanGrant/Programs/TripodsREU/data/uvm_tcga_pan_can_atlas_2018/data_RNA_Seq_v2_expression_median_excel.xlsx',index_col="Hugo_Symbol")
```

```python
expression_df.head()
```

```python
def plot_and_test(expression_df, gene_name,subgroups):
    x = []
    colors= ["b","r","g","y"]
    temp = expression_df.loc[[gene_name]].values[0]
    p = list(expression_df.columns)
    grps = list(subgroups.keys())
    for group in grps:
        x.append([temp[p.index(s[1])] for s in subgroups[group]])
    [sns.kdeplot(x[i],shade=True,color=colors[i],label=grps[i]) for i in range(len(x))]
    for c in combinations(range(len(x)),2):
        print("non-param t-test between ",grps[c[0]],"and ",grps[c[1]],": ", mannwhitneyu(x[c[0]],x[c[1]]))
```

```python
plot_and_test(expression_df,"BAP1", subgroups)
```
