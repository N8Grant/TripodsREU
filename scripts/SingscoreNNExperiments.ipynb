```python
import pandas as pd
import numpy as np
import tensorflow as tf
import TumorDecon as td
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.mixture import BayesianGaussianMixture
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
mixture_df = pd.read_csv("C:/Users/NathanGrant/Programs/TripodsREU/scripts/TumorDecon/data/Simulation_data/Simulation_SNR100to25.csv",index_col="Hugo_Symbol")
mixture_df.head()
```

```python
groundtruth_df = pd.read_csv("C:/Users/NathanGrant/Programs/TripodsREU/scripts/TumorDecon/data/Simulation_data/Simulation_mixture.csv",index_col="Patient")
groundtruth_df.head()
```

```python
lm22 = td.read_lm22_file()
```

```python
def encode_inputs(rna_df,sig_df):
    df_temp = td.df_normalization(rna_df,"minmax")
    rna_df_red ,sig_df_red = td.keep_common_genes(df_temp,sig_df)
    df_temp = rna_df_red.T
    return df_temp
```

```python
input_df = encode_inputs(mixture_df,lm22)
```

```python
ss_results = td.SingScore_main(input_df.T)
```

```python
#ss_results[ss_results<0] = 0
ss_results = ss_results.fillna(0)
del ss_results['Angiogenesis']
del ss_results['Antigen presenting machinery']
ss_results.head()
```

```python
cibersort_results = td.cibersort_main(input_df.T,lm22)
cibersort_results.head()
```

```python
model = tf.keras.Sequential(
    [
        tf.keras.layers.Input(shape=(46)),
        tf.keras.layers.Dense(128,activation="relu"),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(22)
    ]
)
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=opt,
              loss='mae',
              metrics=['mae','mape'])
```

```python
def format_data(cibersort, singscore, groundtruth, test_size=.2,generated_points=0):
    combined_df = pd.concat([cibersort_results,ss_results],axis=1)
    comb = combined_df.values
    ground = groundtruth.values
    indices = np.random.permutation(comb.shape[0])
    training_idx, test_idx = indices[:int(comb.shape[0]*(1-test_size))], indices[int(comb.shape[0]*(test_size)):]
    X_train, X_test,y_train,y_test = comb[training_idx,:], comb[test_idx,:],ground[training_idx,:],ground[test_idx,:]
    if generated_points != 0:
        merged = np.concatenate((X_train,y_train),axis=1)
        bgm = BayesianGaussianMixture(n_components=int(comb.shape[0]*(1-test_size))).fit(merged)
        sampled = bgm.sample(generated_points)
        sampled = sampled[0]
        gen_x = sampled[:,:46]
        gen_y = sampled[:,46:]
        X_train = np.concatenate((X_train,gen_x),axis=0)
        y_train = np.concatenate((y_train,gen_y),axis=0)
    
    return X_train, X_test,y_train,y_test,test_idx
```

```python
X_train, X_test, y_train, y_test,test_index = format_data(cibersort_results,ss_results,groundtruth_df,generated_points=100000)
```

```python
history = model.fit(X_train, y_train, epochs=20, 
                    validation_data=(X_test, y_test))
```

```python
nn_predictions = model.predict(X_test)
c_results = td.cibersort_main(mixture_df.iloc[:,list(test_index)],lm22)
nn_results = pd.DataFrame(data=nn_predictions,columns=c_results.columns)
nn_results = td.sum_upto_1(nn_results)
nn_predictions = nn_results.values
```

```python
print("NN MAE:",mean_absolute_error(y_test,nn_predictions))
print("CIBERSORT MAE:",mean_absolute_error(y_test,c_results.values))
```

```python
lin = LinearRegression().fit(y_test.flatten().reshape(-1, 1),nn_predictions.flatten().reshape(-1, 1))
#plt.scatter(y_test.flatten().reshape(-1, 1),lin.predict(y_test.flatten().reshape(-1, 1)))
plt.scatter(y_test,nn_predictions)
lin.score(y_test.flatten().reshape(-1, 1),nn_predictions.flatten().reshape(-1, 1))**.5
```
